```python
# De breedte van het venster
breedte = 1200

# De hoogte van het venster
hoogte =  480

# De startsnelheid van de strandbal
start_snelheid = [2, 3]

# De startkleur van de achtergrond
start_kleur = [128,0,128]
```

```python
import pygame

screen = None

ball = pygame.image.load("intro_ball.gif")
strandbal = ball.get_rect()

snelheid = start_snelheid
kleur = start_kleur


def bereken_nieuwe_snelheid(vx, vy, links, rechts, onder, boven):
    return vx, vy

def bereken_nieuwe_achtergrond_kleur(huidige_kleur, vx, vy, links, rechts, onder, boven):
    return huidige_kleur

def kleur_scherm(kleur):
    screen.fill(kleur)        
    
def calculate_and_draw_new_frame():
    global strandbal
    global bal
    global snelheid
    global kleur
    
    snelheid = bereken_nieuwe_snelheid(snelheid[0], snelheid[1], strandbal.left, strandbal.right, strandbal.bottom, strandbal.top)
    strandbal = strandbal.move(snelheid)
    
    kleur = bereken_nieuwe_achtergrond_kleur(kleur, snelheid[0], snelheid[1], strandbal.left, strandbal.right, strandbal.bottom, strandbal.top)
    kleur_scherm(kleur)
    
    screen.blit(ball, strandbal)
    pygame.display.flip()

def start():
    global screen
    global strandbal
    
    pygame.init()
    screen = pygame.display.set_mode((breedte, hoogte))


    fps = 120
    FrameUpdateEvent = pygame.USEREVENT+1
    pygame.time.set_timer(FrameUpdateEvent, int(1000/fps))
    
    pygame.event.clear()
    try:
        while True:
            pygame.time.delay(1)
            
            for event in pygame.event.get():
                if event.type == pygame.QUIT: pygame.quit()
                if event.type == FrameUpdateEvent: # is called every 't' milliseconds
                    calculate_and_draw_new_frame()
    except Exception as e:
        ## Uncomment om errors te zien
        # print(e)
        pass
```

```python
# Deel 1
# Tonen wat er gebeurt als ze op start duwen.
# -> Bal verdwijnt aan de rechterkant.

# Zet deze start() in commentaar als dit stuk van de oefening klaar is.
start()
```

```python
# Deel 2
# Zorgen dat de bal niet meer van het scherm verdwijnt.

# Door deze functie te overschrijven kan je bij elke iteratie de snelheid van de bal aanpassen.
# snelheid x en y zijn de snelheid van de bal
# links, rechts, boven en onder zijn de randen van de bal
def bereken_nieuwe_snelheid(snelheid_x, snelheid_y, links, rechts, onder, boven):
    vx = snelheid_x
    vy = snelheid_y

    if rechts > breedte:
        vx = -abs(vx)

    if links < 0:
        vx = abs(vx)

    return vx, vy

# Zet deze start() in commentaar als dit stuk van de oefening klaar is.
start()
```

```python
# Deel 3
# Geef de bal ook een beginsnelheid in de y richting. 
# Zorgen dat de bal niet meer van het scherm verdwijnt.

# Door deze functie te overschrijven kan je bij elke iteratie de snelheid van de bal aanpassen.
# snelheid x en y zijn de snelheid van de bal
# links, rechts, boven en onder zijn de randen van de bal
def bereken_nieuwe_snelheid(snelheid_x, snelheid_y, links, rechts, onder, boven):
    vx = snelheid_x
    vy = snelheid_y

    if rechts > breedte:
        vx = -abs(vx)

    if links < 0:
        vx = abs(vx)

    if onder > hoogte:
        vy = -abs(vy)

    if boven < 0:
        vy = abs(vy)

    return vx, vy

# Zet deze start() in commentaar als dit stuk van de oefening klaar is.
start()
```

```python
# Deel 4
# De achtergrond kleur aanpassen ok te leren met rgb omgaan?
# Pas de achtergrond kleur 

def bereken_nieuwe_achtergrond_kleur(huidige_kleur, vx, vy, links, rechts, onder, boven):
    # Kopieer de huidige kleur
    nieuwe_kleur = [0, 0, 0]
    
    rood = int(min(255, max(0, 255*(links / breedte))))
    groen = int(min(255, max(0, 255*(onder / hoogte))))
    blauw = int(min(255, max(0, (10))))

    nieuwe_kleur[0] = rood
    nieuwe_kleur[1] = groen
    nieuwe_kleur[2] = blauw

    return nieuwe_kleur

start()
```
